```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

### Now i want to normalize the data becaosue somewhere contain the 255 etc

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
from tensorflow.keras.layers import Dropout
model = Sequential()      # now make the model ARchecture
model.add(Flatten(input_shape=(28,28)))   # flatten will make the 2D data into  1D  ( this line of code automatically convert into 1d )
model.add(Dense(256,activation="relu"))   # i want to add 128 hidden layers
Dropout(0.5)
model.add(Dense(128,activation="relu"))   # i want to add 128 hidden layers
Dropout(0.3)
model.add(Dense(64,activation="relu"))   # i want to add 128 hidden layers
Dropout(0.2)
model.add(Dense(10,activation='softmax'))   # always remember when the output is more then two then used activation function (softmax)
```

```python
model.summary()
```

## now to train the model

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
c_object = EarlyStopping(
    monitor='val_loss',  # Watch validation loss
    min_delta=0.00001,   #If the change is less than `min_delta`, it won't count as an improvement. This prevents the training from stopping too early due to tiny fluctuations that don't really matter.
    patience=10,     # Wait 5 epochs after improvement stops( 5 epochs tk imporment nahi ayaa tu stop karigha ye ap set karsakty hai)
    verbose=1,     # it last show the early stopping word just
    #mode='auto',
    restore_best_weights=False  # Keep the best model weights
)
```

```python
model.fit(X_train,y_train,epochs=40,validation_split=0.2,callbacks=c_object)
```

```python
y_prob=model.predict(X_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
